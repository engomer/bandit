```python
import numpy as np
import matplotlib.pyplot as plt
from enum import Enum
from collections import deque
```

```python
def sample_arm(rewardProbabilities, arm):
    return int(float(rewardProbabilities[arm]) > np.random.rand())


def ts(alpha,beta,armToChoose, rewardProbabilities, avgReward, numSelections, gainedReward):
    sampled_means = np.random.beta(alpha, beta)
    selected_arms = np.argsort(sampled_means)[-armToChoose:]
    numSelections[selected_arms] = numSelections[selected_arms] + 1
    for j in selected_arms:
        reward = sample_arm(rewardProbabilities, j)
        alpha[j] = alpha[j] + reward
        beta[j] = beta[j] + 1 - reward
        avgReward[j] = avgReward[j] + (reward - avgReward[j])/numSelections[j]
        gainedReward = gainedReward + reward
    return alpha, beta, avgReward, numSelections, gainedReward

def ts_delayed(alphad, betad, numSelections, avgReward, numOfArms, armToChoose, gainedReward, rewardProbabilities, fifo, fifoSize, delays, delayTime):
    if(not (numSelections > fifoSize).all()):
        selections  = np.random.choice(numOfArms, armToChoose, replace=False)
        while (delays[selections] > 0).any():
            selections  = np.random.choice(numOfArms, armToChoose, replace=False)
        delays[selections] = delayTime
        numSelections[selections] = numSelections[selections] + 1

    else:
        sampled_means = np.random.beta(alphad, betad)
        selections = np.argsort(sampled_means)[-armToChoose:]
        delays[selections] = delayTime
        numSelections[selections] = numSelections[selections] + 1
        for j in selections:
            reward = fifo[j,-1]
            alphad[j] = alphad[j] + reward
            betad[j] = betad[j] + 1 - reward
            avgReward[j] = avgReward[j] + (reward - avgReward[j])/numSelections[j]
            gainedReward = gainedReward + reward
        
    if ((delays==0).any()):
        for m in np.argwhere(delays==0):
            tmp_queue = deque(fifo[m,:].ravel(), maxlen=fifoSize)
            tmp_queue.appendleft(int(sample_arm(rewardProbabilities, m)))
            fifo[m,:] = np.array(tmp_queue)
            delays[m] = -1

    return alphad, betad, numSelections, avgReward, gainedReward, fifo, delays

def ucb_lcb(numSelections, avgReward, numOfArms, armToChoose, roundNum):

    if(not np.all(numSelections)):
        selections  = np.random.choice(numOfArms, armToChoose, replace=False)
    else:
        UCB = avgReward + np.sqrt(2*np.log(roundNum)/numSelections)
        LCB = avgReward - np.sqrt(2*np.log(roundNum)/numSelections)
        selections = np.argpartition(UCB-LCB, -armToChoose)[-armToChoose:]
        
    return selections

def ucb(numSelections, avgReward, numOfArms, armToChoose, roundNum, gainedReward, rewardProbilities):

    if(not np.all(numSelections)):
        selections  = np.random.choice(numOfArms, armToChoose, replace=False)
    else:
        UCB = avgReward + np.sqrt(2*np.log(roundNum)/numSelections)
        selections = np.argpartition(UCB, -armToChoose)[-armToChoose:]
    
    for j in selections:
        reward = sample_arm(rewardProbilities, j)
        numSelections[j] = numSelections[j] + 1
        avgReward[j] = avgReward[j] + (reward - avgReward[j])/numSelections[j]
        gainedReward = gainedReward + reward

    return numSelections, avgReward, gainedReward

def ucb_delayed(numSelections, avgReward, numOfArms, armToChoose, roundNum, gainedReward, rewardProbabilities, fifo, fifoSize, delays, delayTime):

    if(not (numSelections > fifoSize).all()):
        selections  = np.random.choice(numOfArms, armToChoose, replace=False)
        while (delays[selections] > 0).any():
            selections  = np.random.choice(numOfArms, armToChoose, replace=False)
        delays[selections] = delayTime
        numSelections[selections] = numSelections[selections] + 1

        
    else:
        UCB = avgReward + np.sqrt(2*np.log(roundNum)/numSelections)
        selections = np.argpartition(UCB, -armToChoose)[-armToChoose:]
        delays[selections] = delayTime
        numSelections[selections] = numSelections[selections] + 1
        for j in selections:
            reward = fifo[j,-1]
            avgReward[j] = avgReward[j] + (reward - avgReward[j])/numSelections[j]
            gainedReward = gainedReward + reward
    
    if ((delays==0).any()):
        for m in np.argwhere(delays==0):
            tmp_queue = deque(fifo[m,:].ravel(), maxlen=fifoSize)
            tmp_queue.appendleft(sample_arm(rewardProbabilities, m))
            fifo[m,:] = np.array(tmp_queue)
            delays[m] = -1
        

    return numSelections, avgReward, gainedReward, delays, fifo


def perfect(armToChoose, rewardProbabilities, gainedReward, avgReward, numSelections):
    selections  = np.arange(armToChoose)
    for j in selections:
        reward = 1
        gainedReward = gainedReward + reward
        avgReward[j] = avgReward[j] + (reward - avgReward[j])/numSelections[j]
        numSelections[j] = numSelections[j] + 1
    return gainedReward, avgReward, numSelections

def optimal(armToChoose, rewardProbabilities, gainedReward, avgReward, numSelections):
    selections  = np.argsort(rewardProbabilities)[-armToChoose:]
    for j in selections:
        reward = sample_arm(rewardProbabilities, j)
        gainedReward = gainedReward + reward
        avgReward[j] = avgReward[j] + (reward - avgReward[j])/numSelections[j]
        numSelections[j] = numSelections[j] + 1
    return gainedReward, avgReward, numSelections

def roundRobin(armToChoose, numOfArms, rewardProbabilities, gainedReward, avgReward, numSelections, roundNum):
    selections  = (np.arange(armToChoose) + roundNum % numOfArms) % numOfArms
    for j in selections:
        reward = sample_arm(rewardProbabilities, j)
        gainedReward = gainedReward + reward
        avgReward[j] = avgReward[j] + (reward - avgReward[j])/numSelections[j]
        numSelections[j] = numSelections[j] + 1
    return gainedReward, avgReward, numSelections

class Algos(Enum):
    UCB = 0
    UCB_DELAYED_5 = 1
    TS = 2
    TS_DELAYED_5 = 3
    OPTIMAL = 4
    UCB_DELAYED_10 = 5
    TS_DELAYED_10 = 6
    PERFECT = 7
    ROUND_ROBIN = 8
```

```python
numOfArms = 100
armToChoose = 3
numOfRounds = 25000

algoNum = len(Algos)
fifoSize = 10
fifo = np.zeros((algoNum, numOfArms, fifoSize))

rewardProbabilities = np.random.uniform(0,0.3,(numOfArms))
rewardProbabilities[0] = 0.9
rewardProbabilities[1] = 0.95
rewardProbabilities[2] = 0.8

avgRewardArray = np.zeros((algoNum, numOfRounds))
numSelections = np.zeros((algoNum, numOfArms))
avgReward = np.zeros((algoNum, numOfArms))
delays = np.zeros((algoNum, numOfArms)) -1

gainedReward = np.zeros(algoNum)


alpha = np.ones(numOfArms)
alphad = np.ones(numOfArms)

beta = np.ones(numOfArms)
betad = np.ones(numOfArms)

DELAYY = 5

for i in range(1,numOfRounds):
    # UCB
    numSelections[Algos.UCB.value,:],  avgReward[Algos.UCB.value,:], gainedReward[Algos.UCB.value]= ucb(numSelections[Algos.UCB.value, :], avgReward[Algos.UCB.value, :], numOfArms, armToChoose, i, gainedReward[Algos.UCB.value], rewardProbabilities)
    avgRewardArray[Algos.UCB.value,i] = gainedReward[Algos.UCB.value] / (i*numOfArms)
    # UCB Delayed 5
    numSelections[Algos.UCB_DELAYED_5.value,:],  avgReward[Algos.UCB_DELAYED_5.value,:], gainedReward[Algos.UCB_DELAYED_5.value], delays[Algos.UCB_DELAYED_5.value,:], fifo[Algos.UCB_DELAYED_5.value, :, :] = ucb_delayed(numSelections[Algos.UCB_DELAYED_5.value, :], avgReward[Algos.UCB_DELAYED_5.value, :], numOfArms, armToChoose, i, gainedReward[Algos.UCB_DELAYED_5.value], rewardProbabilities, fifo[Algos.UCB_DELAYED_5.value,:,:], fifoSize, delays[Algos.UCB_DELAYED_5.value,:], 5)
    avgRewardArray[Algos.UCB_DELAYED_5.value,i] = gainedReward[Algos.UCB_DELAYED_5.value] / (i*numOfArms)
    # UCB Delayed 10
    numSelections[Algos.UCB_DELAYED_10.value,:],  avgReward[Algos.UCB_DELAYED_10.value,:], gainedReward[Algos.UCB_DELAYED_10.value], delays[Algos.UCB_DELAYED_10.value,:], fifo[Algos.UCB_DELAYED_10.value, :, :] = ucb_delayed(numSelections[Algos.UCB_DELAYED_10.value, :], avgReward[Algos.UCB_DELAYED_10.value, :], numOfArms, armToChoose, i, gainedReward[Algos.UCB_DELAYED_10.value], rewardProbabilities, fifo[Algos.UCB_DELAYED_10.value,:,:], fifoSize, delays[Algos.UCB_DELAYED_10.value,:], 10)
    avgRewardArray[Algos.UCB_DELAYED_10.value,i] = gainedReward[Algos.UCB_DELAYED_10.value] / (i*numOfArms)
    # TS
    alpha, beta, avgReward[Algos.TS.value,:], numSelections[Algos.TS.value,:], gainedReward[Algos.TS.value] = ts(alpha, beta, armToChoose, rewardProbabilities, avgReward[Algos.TS.value,:], numSelections[Algos.TS.value,:], gainedReward[Algos.TS.value])
    avgRewardArray[Algos.TS.value,i] = gainedReward[Algos.TS.value] / (i*numOfArms)
    # TS Delayed 5
    alphad, betad, numSelections[Algos.TS_DELAYED_5.value,:], avgReward[Algos.TS_DELAYED_5.value,:], gainedReward[Algos.TS_DELAYED_5.value], fifo[Algos.TS_DELAYED_5.value, :, :], delays[Algos.TS_DELAYED_5.value,:] = ts_delayed(alphad, betad, numSelections[Algos.TS_DELAYED_5.value,:], avgReward[Algos.TS_DELAYED_5.value,:], numOfArms, armToChoose, gainedReward[Algos.TS_DELAYED_5.value], rewardProbabilities, fifo[Algos.TS_DELAYED_5.value,:,:], fifoSize, delays[Algos.TS_DELAYED_5.value,:], DELAYY)
    avgRewardArray[Algos.TS_DELAYED_5.value,i] = gainedReward[Algos.TS_DELAYED_5.value] / (i*numOfArms) 
    # TS Delayed 10
    alphad, betad, numSelections[Algos.TS_DELAYED_10.value,:], avgReward[Algos.TS_DELAYED_10.value,:], gainedReward[Algos.TS_DELAYED_10.value], fifo[Algos.TS_DELAYED_10.value, :, :], delays[Algos.TS_DELAYED_10.value,:] = ts_delayed(alphad, betad, numSelections[Algos.TS_DELAYED_10.value,:], avgReward[Algos.TS_DELAYED_10.value,:], numOfArms, armToChoose, gainedReward[Algos.TS_DELAYED_10.value], rewardProbabilities, fifo[Algos.TS_DELAYED_10.value,:,:], fifoSize, delays[Algos.TS_DELAYED_10.value,:], 10)
    avgRewardArray[Algos.TS_DELAYED_10.value,i] = gainedReward[Algos.TS_DELAYED_10.value] / (i*numOfArms)
    # Optimal
    gainedReward[Algos.OPTIMAL.value], avgReward[Algos.OPTIMAL.value,:], numSelections[Algos.OPTIMAL.value,:] = optimal(armToChoose, rewardProbabilities, gainedReward[Algos.OPTIMAL.value], avgReward[Algos.OPTIMAL.value,:], numSelections[Algos.OPTIMAL.value,:])
    avgRewardArray[Algos.OPTIMAL.value,i] = gainedReward[Algos.OPTIMAL.value] / (i*numOfArms)
    # Perfect
    gainedReward[Algos.PERFECT.value], avgReward[Algos.PERFECT.value,:], numSelections[Algos.PERFECT.value,:] = perfect(armToChoose, rewardProbabilities, gainedReward[Algos.PERFECT.value], avgReward[Algos.PERFECT.value,:], numSelections[Algos.PERFECT.value,:])
    avgRewardArray[Algos.PERFECT.value,i] = gainedReward[Algos.PERFECT.value] / (i*numOfArms)
    # Round Robin
    gainedReward[Algos.ROUND_ROBIN.value], avgReward[Algos.ROUND_ROBIN.value,:], numSelections[Algos.ROUND_ROBIN.value,:] = roundRobin(armToChoose, numOfArms, rewardProbabilities, gainedReward[Algos.ROUND_ROBIN.value], avgReward[Algos.ROUND_ROBIN.value,:], numSelections[Algos.ROUND_ROBIN.value,:], i)
    avgRewardArray[Algos.ROUND_ROBIN.value,i] = gainedReward[Algos.ROUND_ROBIN.value] / (i*numOfArms)


    delays[delays > 0] = delays[delays > 0] - 1
    
#print(numSelections[Algos.UCB.value,:])
#print(avgRewardArray[Algos.UCB.value,:])
#print("\n")
#print("UCB")
#print(numSelections[Algos.UCB.value,:])
#print(avgRewardArray[Algos.UCB.value,:])
#print("\n")
#print("UCB Delayed")
#print(numSelections[Algos.UCB_DELAYED_5.value,:])
#print(avgRewardArray[Algos.UCB_DELAYED_5.value,:])
#print("\n")
#print("TS")
#print(numSelections[Algos.TS.value,:])
#print(avgRewardArray[Algos.TS.value,:])
#print("\n")
#print("TS Delayed")
#print(numSelections[Algos.TS_DELAYED_5.value,:])
#print(avgRewardArray[Algos.TS_DELAYED_5.value,:])
#print("\n")
#print("Optimal")
#print(numSelections[Algos.OPTIMAL.value,:])
#print(avgRewardArray[Algos.OPTIMAL.value,:])
#print("\n")
#print("Perfect")
#print(numSelections[Algos.PERFECT.value,:])
#print(avgRewardArray[Algos.PERFECT.value,:])
#print("\n")
#print("Round Robin")
#print(numSelections[Algos.ROUND_ROBIN.value,:])
#print(avgRewardArray[Algos.ROUND_ROBIN.value,:])
#print("\n")

plt.plot(avgRewardArray[Algos.UCB.value,:], label="UCB")
plt.plot(avgRewardArray[Algos.UCB_DELAYED_5.value,:], label="UCB Delayed 5")
plt.plot(avgRewardArray[Algos.UCB_DELAYED_10.value,:], label="UCB Delayed 10")
plt.plot(avgRewardArray[Algos.TS.value,:], label="TS")
plt.plot(avgRewardArray[Algos.TS_DELAYED_5.value,:], label="TS Delayed 5")
plt.plot(avgRewardArray[Algos.TS_DELAYED_10.value,:], label="TS Delayed 10")
plt.plot(avgRewardArray[Algos.OPTIMAL.value,:], label="Optimal")
plt.plot(avgRewardArray[Algos.PERFECT.value,:], label="Perfect")
plt.plot(avgRewardArray[Algos.ROUND_ROBIN.value,:], label="Round Robin")

plt.legend()
plt.show()
plt.tight_layout()
```
